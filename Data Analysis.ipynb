```python
fp=pd.read_csv("C:\\Users\\91957\\Documents\\AIT580\\Crash_Reporting_-_Drivers_Data.csv")
```

```python
fp.info()
```

```python
#categorical variables
fp_cat=fp.drop(columns=['Latitude','Longitude','Vehicle Year','Speed Limit','Location'],axis=1)
```

```python
columns_cat=['Route Type', 'Road Name','Cross-Street Type','Cross-Street Name','Off-Road Description','Municipality','Related Non-Motorist','Drivers License State','Vehicle Make', 'Vehicle Model']           
for col in columns_cat:
    fp_cat[col].fillna('UNKNOWN',inplace=True)
    print(fp_cat[col].isna().sum())
```

```python
columns_cat1=['Collision Type','Weather','Surface Condition','Light','Traffic Control','Driver Substance Abuse','Non-Motorist Substance Abuse','Vehicle Damage Extent','Vehicle First Impact Location','Vehicle Second Impact Location','Vehicle Body Type','Vehicle Movement','Vehicle Continuing Dir','Vehicle Going Dir','Equipment Problems']
for col in columns_cat1:
    statmode=statistics.mode(fp_cat[col])
    fp_cat[col].fillna(statmode,inplace=True)
    print(fp_cat[col].isna().sum())
```

```python
#numerical variables
fp_num=fp.drop(columns=fp_cat.columns,axis=1)
```

```python
fp_cat.drop('Non-Motorist Substance Abuse',axis=1,inplace=True)
fp_cat.drop('Circumstance',axis=1,inplace=True)
```

```python
fp_cat.info()
```

```python
col_num=['Latitude','Longitude','Vehicle Year','Speed Limit']
for col in col_num:
    statmean=np.mean(fp_num[col])
    fp_num[col].fillna(statmean,inplace=True)
    print(fp_num[col].isna().sum())
```

```python
new_data=pd.concat([fp_cat,fp_num],axis=1)
new_data.info()
```

```python
new_data.drop('Crash Date/Time',axis=1,inplace=True)
```

```python
# Convert the 'DateTime' column to datetime format
fp['Crash Date/Time'] = pd.to_datetime(fp['Crash Date/Time'], errors='coerce')
fp['Crash Date/Time'].fillna('2023-01-01 00:00:00',inplace=True)
print(fp['Crash Date/Time'].head(10))
print(fp['Crash Date/Time'].isna().sum())
```

```python
df1 = new_data.reset_index(drop=True)
df2 = fp['Crash Date/Time'].reset_index(drop=True)
result_df = pd.concat([df1, df2], axis=1)
```

```python
result_df.info()
```
